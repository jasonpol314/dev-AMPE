# Lectura del fichero y creacion de una estructura de datos

- leemos del fichero origen
- saltamos las lieneas vacias
- y creamos un registro (diccionario) por cada persona

```python
def sacar_nombre(l):

    if 'Origen:' in l:
        pri=l.index('Origen:')
        fin=l.index('Puntuac')
        nombre=[]
        for i in range(pri+1,fin):
            nombre.append(l[i])
 
        longi=len(nombre)
        origen= nombre[longi-2]+nombre[longi-1]
        
        nombrepl=nombre[:longi-2]
        nombreps=''
        for i in nombrepl:
            nombreps +=' '
            nombreps += i
        nombreps=nombreps.strip()
        
        laux=nombreps.split(',')
        lapellidos=laux[0].split('  ')
        lnombre=laux[1].split('  ')
        
        apellidos=''
        for ap in lapellidos:
            apellidos += ' '
            apellidos += ap.replace(' ','')
        apellidos=apellidos.strip()
            
        nombre = ''
        for nom in lnombre:
            nombre +=' '
            nombre += nom.replace(' ','')
        nombre=nombre.strip()
        
        nombretot= apellidos+', '+nombre
        return(nombretot, apellidos, nombre, origen)
    else:
        return (None, None, None, None)

def sacar_puntuacion(l):
    if 'ión:' in l:
        k=l.index('ión:')
        return l[k+1]
    else:
        return None

def sacar_destino(l):  
    for x in l:
        if len(x)>8 and x[:8] == 'Destino:':
            dest= x[9:].replace('\t', ' ')
            return dest
    else:
        return None

def sacar_peticion(l):  
    for x in l:
        if len(x)>11 and x[:11] == 'Adjudicada:':
            return(x[12:])
        else:
            if len(x)>20 and x[:20] == 'Petición Adjudicada:':
                return (x[20:])
    else:
        return None
    
def parsear_centro(s):
    l=s.split()
    
    cp=l[0]
    if cp.isnumeric():
        pass
    else:
        cp= 0
        l = ['0']+l
    
    lde= s.split('DE ')
    long= len(lde)
    if long>1:
        localidad= lde[long-1].strip()
    else:
        localidad= 'Ninguna'
        
    lcentro=list(reversed(l[1:]))
    
    centro=''
    if localidad == 'Ninguna':
        quitar_centro= False
    else:
        quitar_centro= True
        
    for i in lcentro:
        if i == 'DE':
            quitar_centro=False
        else:
            if quitar_centro:
                pass
            else:
                centro =  i+ ' '+ centro    

    return (s, cp, centro, localidad) 
    
    
def sacar_centro(l):
    for x in l:
        if len(x)>7 and x[:7] == 'Centro:':
            return parsear_centro(x[8:])
            
    else:
        return (None, None, None, None)
    
def parsear_especialidad(s):
    l1=s.split('(V)')
    l2=s.split('(C)')
    
    if len(l1)>1:
        espe1= l1[0]
        espe2= l1[1]
        idioma= 'Euskera'
    elif len(l2)>1:
        espe1= l2[0]
        espe2= l2[1]
        idioma= 'Castellano'
    else:
        espe1=l1[0]
        espe2= ''
        idioma='Ninguno'
        
    espe2=espe2.replace('-','').strip()
    if espe2 == '': espe2 = 'Ninguna' 
        
    return (espe1, idioma, espe2)

def sacar_especialidad2(l):
    aux= 'Especialidad:\t'
    long= len(aux)
    for x in l:
        if len(x)>long and x[:long] == aux:
            espe_total=x[long:]
            (e1, id, e2)= parsear_especialidad(espe_total)
            return(e1, id, e2)
    else:
        return (None, None, None)
    
```

```python
def crear_registro(s):
    ls = s.split('\n')
    (nombretot, apellidos, nombre, origen)= sacar_nombre(ls)
    punt= sacar_puntuacion(ls)
    dest= sacar_destino(ls)
    peti= sacar_peticion(ls)
    (cent,cp, centro, localidad)= sacar_centro(ls)
    (espe1, idio, espe2) = sacar_especialidad2(ls)
    dic = {'nombre completo': nombretot,
          'apellidos': apellidos,
          'nombre': nombre,
          'origen': origen,
          'puntuacion': punt,
          'destino': dest,
          'peticion': peti,
          'centro': cent,
          'cp':cp,
          'nombre centro':centro,
          'localidad': localidad,
          'especialidad 1': espe1,
          'idioma': idio,
          'especialidad 2': espe2}
    return (ls,dic)
```

```python
def imprimir_registro(dic):
    long=35
    relleno= '-'* long
    
    for clave in dic:
        k=len(clave)
        rell= relleno[:long-k]
        print('{c} {r}> {v}'.format(c=clave, r=rell, v=dic[clave]))
```

```python
import csv

datainput= ''
cont=0
contlin= 0

with open('2021Adjudicacion.txt', mode='rt') as f:
    with open('names2021.csv', 'w', newline='', encoding='latin-1') as csvfile:
        fieldnames = ['nombre completo',
              'apellidos',
              'nombre',
              'origen',
              'puntuacion',
              'destino',
              'peticion',
              'centro',
              'cp',
              'nombre centro',
              'localidad',
              'especialidad 1',
              'idioma',
              'especialidad 2']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        
        for line in f:
        
            contlin+=1
            line=line[:-1]
            cline=line
        
            if cline == '': continue
            
            if cline == 'Nombre: ':
                (linea,d) = crear_registro(datainput)
                print(linea)
                print('\n')
                imprimir_registro(d)
                print('\n\n')
                writer.writerow(d)
                
                datainput = cline
                cont+=1
            else:
                cline='\n'+cline
                datainput += cline
        (linea,d) = crear_registro(datainput)
        imprimir_registro(d)
        writer.writerow(d)
    
    print(cont)
    print(contlin)
```
